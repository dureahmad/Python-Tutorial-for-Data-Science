#  List
A list is a data structure in Python that is a **mutable**, or changeable. Each element or value that is inside of a list is called an item. Just as strings are defined as characters between quotes, lists are defined by having values between **square brackets [], or by using list(----).**

```python
my_list = ["Lesson", 5, "Is Fun?", True]
print(my_list)
my_list.remove(5)
print(my_list)
```

```python
empty_list = []
print(empty_list)
print('------\n')
empty_list.append('I am no longer empty')
print(empty_list)
print('------\n')
list_1 = list('I am a data scientist')
print(list_1)
```

Note that we can use

    * append(), remove(), extend(), sort(), insert(), reverse(),  len(). max(), min()

and many more commands on list

```python
list_2 = list('Muhammad')
print(list_2, '\n')

list_2.remove('M')
print(list_2, '\n'*3)   # create 3 new lines

# To append some letter or world at the end
list_2.append('K')  # It will append K at the end (one at a time)
list_2.append('Kudo') 
print(list_2)
```

If we want to add in the beginning, we need to define new list and extend with the old one using extend() command.

```python
#
list_3 = list('M')    # we can also define this like lisr_3 = ['K']
list_3.extend(list_2)
print(list_3)
```

```python
num_list = [1, 3, 5, 7, 9]
print( 'Lnegth', len(num_list))             # Check the length
print( 'Max', max(num_list))                # Check the max
print( 'Min', min(num_list))                # Check the min
print( 'sum', sum(num_list))                # Check the sum
print( 'Mean',sum(num_list)/len(num_list))  # Check the mean*

print('Number of time 3 appears is -----', num_list.count(3))
```

More Examples of List

```python
# Examples
x = range(10)   # is the list [0, 1, ..., 9]
zero = x[0]     # equals 0, lists are 0-indexed
one = x[1]   # equals 1
nine = x[-1] # equals 9, 'Pythonic' for last element
eight = x[-2] # equals 8, 'Pythonic' for next-to-last element
len(x)        # 10


# You can also use square brackets to “slice” lists:

first_three = x[:3]          # From beginning upto the thrid in list, i.e., [0, 1, 2]
three_to_end = x[3:]         # From 3rd in the list upto the end of list, i.e., [3, 4, ..., 9]
one_to_four = x[1:5]         # [1, 2, 3, 4]
last_three = x[-3:]          # [7, 8, 9]
without_first_and_last = x[1:-1]   # [1, 2, ..., 8]
copy_of_x = x[:]                   # [-1, 1, 2, ..., 9]
print(list(x))
list(x[:5:-1])   #[start:end:step])
list(x[:6:2])
```

# Note about Link List
A linked list is a sequence of data elements, which are connected together via links. Each data element contains a connection to another data element in form of a pointer. Python does not have linked lists in its standard library.
We need to use 'Class' to create one.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

# Tuples
Tuples are lists’ immutable cousins. Pretty much anything you can do to a list that doesn’t
involve modifying it, you can do to a tuple. You specify a tuple by using parentheses (or
nothing) instead of square brackets:


```python
my_list = [1, 2]           # it is equalent to my_list = list(1,2)
my_tuple = (1, 2)
other_tuple = 3, 4
my_list[1] = 3             # my_list is now [1, 3]
```

Tuples are a convenient way to return multiple values from functions:

```python
def sum_and_product(x, y):
    return (x + y),(x * y)

sp = sum_and_product(2, 3) # equals (5, 6)
s, p = sum_and_product(5, 10) # s is 15, p is 50
```

Tuples (and lists) can also be used for multiple assignment:

```python
x, y = 1, 2 # now x is 1, y is 2
x, y = y, x # Pythonic way to swap variables; now x is 2, y is 1
```

# Dictionaries
Another fundamental data structure is a dictionary, which associates values with keys and
allows you to quickly retrieve the value corresponding to a given key: 
we use { } brackets for dictionary


```python
empty_dict = {} # Pythonic
empty_dict2 = dict() # less Pythonic
grades = { "Joel" : 80, "Tim" : 95 } # dictionary literal

# You can look up the value for a key using square brackets:

print('Grades of Joel is  ',grades['Joel'])   # equals 80

print(grades)

```

You can check for the existence of a key using in:



```python
joel_has_grade = "Joel" in grades # True
kate_has_grade = "Kate" in grades # False

# adds a new entry
grades["Kate"] = 100

len(grades)   #Now it is 3

```

We will frequently use dictionaries as a simple way to represent structured data: We can use the commands like
* key(), values(), items()



```python
tweet = {"user" : "joelgrus",
"text" : "Data Science is Awesome",
"retweet_count" : 100,
"hashtags" : ["#data", "#science", "#datascience", "#awesome", "#yolo"]
}

tweet_keys = tweet.keys() # list of keys ['user', 'text', 'retweet_count', 'hashtags']

tweet_values = tweet.values() # list of values ['joelgrus', 'Data Science is Awesome', 100, ['#data', '#science', '#datascience', '#awesome', '#yolo']]

tweet_items = tweet.items() # list of (key, value) tuples



```

# Strings
Strings can be delimited by single or double quotation marks. You can create multiline strings using triple-[double-]-quotes:

```python
s1 = "hello! How are you. looking forward to work as data scientist Inshallah."
s2 = """hello! 
How are you. 
looking frorward to work on python."""
print(s1)
print('--------------')
print(s2)
```

```python
print(s1[0:5])     # print hello
print(s1[-10])     # print I, the first letter of Inshallah from the end
print(s1[-10:])    # from the position -1 to end (Inshallah)
print(s1[:-10])    # print upto the last 10th character

print(s1[:-10]+ 'starting from next month  ' + s1[-10:])
```

```python
print('%s is my %s letter and %s is my number %.5f ' %('M', 'favourite', '7/2', 7/2))
```

We can use various commands with strings such as

* capatilize(), find(), replace(), split() and more

```python
print(s1.capitalize())   # make the first letter capital
print(s1.find('work'))   # Find the location of the word 'work'
print(len(s1))           # print length of the string

print(s1.replace('work in','join'))   # will replace 'work in' with 'join'

print(s1.split(" "))
```

# Arrays

```python
# another example
def cm(feet =0, inches = 0):
    '''Convert length from feet and inches to cm'''  
    inch_cm = inches*2.54
    feet_cm = feet*12*2.54
    
    return inch_cm + feet_cm

print('feet to cm:',cm(feet=5))
print('inhes to cm:', cm(inches = 2))
print('feet and inches to cm:  %.3f' %cm(feet = 5, inches = 2))
```

# Lambda Expression

lambda expressions or anonymous functions are function definitions that are not bound to an identifier. They are often used as arguments being passed to or values returned from higher-order functions

https://www.youtube.com/watch?v=25ovCm9jKfA

```python
# Standard way
def f(n):
    return 3*n+1

print('f(5) =  ', f(5))

# Now we use lambda expression
g = lambda x: 3*x+1
print('g(5) =  ',g(5))
```

```python
def y1(x):
    return(x+4)
print(y1(5))

# Now using lambda function
y1=lambda x:x+4
y1(9)
```

```python
def subt(a,b):
    return(a-b)
print(subt(5,11))

# Lambda Way
y2=(lambda a,b: a-b)
y2(5,15)
```

```python
# Quadratic function using lambda function
def qf(a,b,c):
    return lambda x: a*x**2+b*x+c

f=qf(2,-1,3)  # qf(a,b,c)
print('f(x) = ax^2+bx+c --------f(5) = ', f(5))
```

```python
def fizzBuzz(n):
   
    if n%3==0 and n%5==0:
        print('FizzBuzz')
    elif n%3==0 and n%5!=0:
        print('Fizz')
    elif n%5==0 and n%3!=0:
        print('Buzz')
    else:
        print(n)

n = int(input().strip())
fizzBuzz(n)
```
